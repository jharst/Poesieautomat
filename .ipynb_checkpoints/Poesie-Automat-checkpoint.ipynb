Diese Maschine stellt eine dreidimensionale Liste dar (sie besteht aus sechs Zeilen, die je sechs Spalten haben; jede Spalte versammelt sechs mögliche Ausdrücke). Ich beschränke mich der Übersichtlichkeit halber auf eine einzige Zeile, die als zweidimensionale Liste dargestellt werden kann (beschränkt auf je drei Elemente):

|1|2|3|4|5|6|
|-|-|-|-|-|-|
|Brühwarme|Andacht|zum Nettopreis|Dieser massive|Mundgeruch|am Wochenende|
|Noch und noch|Hoffnung|zum Frühstück.|Dieser beklommene|Partnertausch|vor dem Sturm.|
|Ein Eßlöffel|Finsternis|am rechten Fleck.|Dieser staubige|Todeswunsch|in hohem Alter.|


Die Umsetzung des Automaten könnte dann wie folgt funktionieren:
```mermaid
flowchart TD
    A(Führe für jede Zeile der Liste aus:) --> B(Wähle ein zufälliges Element der nächsten Zeile.)
    B --> C(Füge es dem String 'verse' hinzu.)
    C --> E{Fertig?}
    E -- ja --> D(Gib den String 'verse' aus.)
    E -- nein --> B
```


Ich beginne mit der Definition der Daten in einem zweidimensionalen Array (ein Vers mit sechs Elementen setzt eine Liste von sechs Listen voraus). Ich importiere das Modul "random" zur Generation von zufälligen Auswahlmöglichkeiten und initialisiere den Datensatz.

````python
from random import choice, shuffle

verse_elements = [
    ['Brühwarme', 'Noch und noch', 'Ein Eßlöffel'],
    ['Andacht', 'Hoffnung', 'Finsternis'],
    ['zum Nettopreis.', 'zum Frühstück.', 'am rechten Fleck.'],
    ['Dieser massive', 'Dieser beklommene', 'Dieser staubige'],
    ['Mundgeruch', 'Partnertausch', 'Todeswunsch'],
    ['am Wochenende.', 'vor dem Sturm.', 'in hohem Alter.']
]
````

Anschließend schreibe ich eine Funktion, die einen Vers aus den Daten per Zufallsprinzip zusammensetzt:

````python
def generate_verse(elements):
    verse = ""
    for element in elements:
        verse += choice(element) + " "
    return verse.strip()
````

Und zuletzt rufe ich die Funktion in einem print-statement auf:

````python
print(generate_verse(verse_elements))
````
